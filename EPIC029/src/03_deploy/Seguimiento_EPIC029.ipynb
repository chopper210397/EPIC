```python
dataframe['PERSONA_BEN']=0
dataframe.loc[(dataframe['DESBANCA_BENEFICIARIO']== 'BANCA PERSONAL                     '),'PERSONA_BEN']=1

dataframe['EMPRESA_BEN']=0
dataframe.loc[(dataframe['DESBANCA_BENEFICIARIO']== 'EMPRESA                            '),'EMPRESA_BEN']=1

dataframe["FLG_NOREL"]=np.where(dataframe["FLG_REL"]==1,0,1)
```

```python
columnas_numericas = ['MTOTRANSACCION']
columnas_categoricas = ['PERSONA_BEN','EMPRESA_BEN','FLG_NOREL','FLG_ROS_BEN','FLG_ROS_SOL']
name_colum_periodo = ['PERIODO']
```

```python
df_0 = dataframe.sample(frac = 0.7,random_state=7)
df_t = dataframe.drop(df_0.index)
df_0.reset_index(drop=True)
df_t.reset_index(drop=True)
```

```python
df_t['PERIODO']=202010
```

```python
def stability_function(df_0,df_t,columnas_numericas,columnas_categoricas,name_colum_periodo):
    #df : dataframe
    #columnas_numericas : variables continuas
    #columnas_categoricas : variables categóricas, flags
    #name_colum_periodo : nombre de la columna del periodo a evaluar
    rows_flg = []
    rows_psi = []
    rows = []
    rows2 = []
    rows3 = []
    rows4 = []
    rows5 = []
    columnas_total=columnas_numericas+columnas_categoricas
    for i in columnas_total:
        print(i)
        A = df_0[[i]]
        B = df_t[[i]] #poner aqui el df de la base de miguel 
        TotalRowsA = A.shape[0]
        TotalRowsB = B.shape[0]
        MissingA=A[i].isnull().sum()/TotalRowsA
        MissingB=B[i].isnull().sum()/TotalRowsB
        if i in columnas_numericas:
            A['corte'],bins  = pd.qcut(A[i], q = 5,retbins=True, duplicates='drop')       
            bins[0] = -math.inf
            bins[len(bins)-1] = math.inf
            A['corte'], h  = pd.cut(A[i], bins = bins,retbins = True)
            B['corte'], h  = pd.cut(B[i], bins = bins,retbins = True)
            A = (A['corte'].value_counts()/ TotalRowsA).to_frame().sort_index()
            B = (B['corte'].value_counts()/ TotalRowsB).to_frame().sort_index()
            values_1 = list(A['corte'].values)
            values_2 = list(B['corte'].values)
        else :
            A = (A[i].value_counts()/ TotalRowsA).to_frame().sort_index()
            B = (B[i].value_counts()/ TotalRowsB).to_frame().sort_index()
            C = pd.concat([A,B], axis= 1)
            B = C.iloc[: , 1:].fillna(0)
            values_1 = list(A[i].values)
            values_2 = list(B[i].values)
            
        categorias = list(A.index.values)
        rows2.append('Missing')
        rows3.append(MissingA)
        rows4.append(MissingB)
        if i in columnas_numericas:
            for h in range(len(categorias)):
                rows2.append(('('+str(categorias[h].left)+','+str(categorias[h].right)+')'))
        else :
            for j in categorias:
                rows2.append(j)
        for j in values_1:
            rows3.append(j)
        for j in values_2:
            rows4.append(j)
        
        PSI = (A-B)*np.log(A/B)
        PSI_i=PSI.replace([np.inf, -np.inf],0.2)
        PSI = PSI_i.sum(axis = 0, skipna = True)
        PSI = PSI[0]
        rows.append(PSI)
        
        for j in categorias:
            rows_flg.append(i)
            rows_psi.append(PSI)
        rows_flg.append(i)
        rows_psi.append(PSI)
        #rows5=rows5+PSI_i.iloc[:,0].tolist()
    periodo=df_t[name_colum_periodo][0]
    CATEGORIAS_DF = pd.DataFrame({'PERIODO':periodo,
                                  'DISTRIBUCIÓN MODELAMIENTO':rows3,
                                  'DISTRIBUCIÓN ACTUAL':rows4,
                                  'CATEGORIAS':rows2,
                                  'VARIABLES':rows_flg,
                                  #'PSI_i':rows5,
                                  'PSI':rows_psi})
    
    conditions = [
    (CATEGORIAS_DF['PSI'] < 0.1),
    (CATEGORIAS_DF['PSI'] >= 0.1) & (CATEGORIAS_DF['PSI'] < 0.2),
    (CATEGORIAS_DF['PSI'] >= 0.2)
    ]

    # create a list of the values we want to assign for each condition
    values = ['ESTABLE', 'LIGERA INESTABILIDAD', 'INESTABLE']

    # create a new column and use np.select to assign values to it using our lists as arguments
    CATEGORIAS_DF['RESULTADO'] = np.select(conditions, values)
    CATEGORIAS_DF=CATEGORIAS_DF[['PERIODO','VARIABLES','CATEGORIAS','DISTRIBUCIÓN MODELAMIENTO','DISTRIBUCIÓN ACTUAL',
                                 #'PSI_i',
                                 'PSI','RESULTADO']]
    return(CATEGORIAS_DF,periodo)
```

```python
salida, periodo = stability_function(df_0,df_t,columnas_numericas,columnas_categoricas,'PERIODO')
```

```python
salida.to_excel("RESULTADOS_"+str(periodo)+".xlsx")
```
